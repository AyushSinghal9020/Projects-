# 4 | Model Building

````python
train , test = np.split(data.sample(frac = 1) , [int(0.8 * len(data))])
````

````python
def pre(dataframe):
    x = dataframe.drop("Age" , axis = 1)
    y = dataframe["Age"]
    
    return x , y
````

````python
X_train , y_train = pre(train)
X_test , y_test = pre(test)
````
